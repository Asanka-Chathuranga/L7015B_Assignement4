```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import SelectKBest, f_classif, RFE
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Load the three csv files
file1 = r"C:\Users\user\OneDrive - Luleå University of Technology\TMUTA 2024-2025\D7015B- Industrial AI and eMaintenance\Assignment_4\Trail1_extracted_features_acceleration_m1ai1.csv"
file2 = r"C:\Users\user\OneDrive - Luleå University of Technology\TMUTA 2024-2025\D7015B- Industrial AI and eMaintenance\Assignment_4\Trail2_extracted_features_acceleration_m1ai1.csv"
file3 = r"C:\Users\user\OneDrive - Luleå University of Technology\TMUTA 2024-2025\D7015B- Industrial AI and eMaintenance\Assignment_4\Trail3_extracted_features_acceleration_m2ai0.csv"

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)  # This has 2 additional columns

# Combine all three files, aligning columns
df = pd.concat([df1, df2, df3], ignore_index=True)

# Handling missing values for the extra columns in df3
df = df.dropna() # Replace NaN with mean value as a strategy

# Save the final combined table
df.to_csv(r"C:\Users\user\OneDrive - Luleå University of Technology\TMUTA 2024-2025\D7015B- Industrial AI and eMaintenance\Assignment_4\combined_output.csv", index=False)

#print("Merged file saved as combined_output.xlsx")
```

```python
df = df.drop(columns=["start_time", "axle", "cluster", "tsne_1", "tsne_2"])
df.head()
```

```python
df["event"] = df["event"].replace({"normal": 0})  # Replace 'normal' with 0
df["event"] = df["event"].apply(lambda x: 1 if x != 0 else x)  # Replace everything else with 1
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Identify the feature columns (exclude the 'event' column)
feature_columns = df.columns.drop('event')

# Apply Z-score normalization only to the feature columns
df[feature_columns] = scaler.fit_transform(df[feature_columns])

# Display the first few rows of the normalized DataFrame
df.head()
```

```python
# Extract features (normalized data) and target
X = df[feature_columns]  # Actual normalized feature data
y = df['event']          # Target label

# Split the data to 80% /20%
X_train, X_test, y_train, y_test = train_test_split(
    X,                   # Use the normalized feature data (not just column names)
    y,
    test_size=0.20,
    random_state=42,
    stratify=y           # Requires y to be a categorical/class label
)
```

```python
# Verifying the split
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("Training events distribution:\n", y_train.value_counts(normalize=True))
print("Testing events distribution:\n", y_test.value_counts(normalize=True))
```

```python
# Filtering with Pearson Correlation
def pearson_correlation_filter(df, target_col, threshold=0.3):
    correlation_matrix = df.corr()[target_col].abs()
    selected_features = correlation_matrix[correlation_matrix > threshold].index.tolist()
    selected_features.remove(target_col)
    return selected_features

pearson_features = pearson_correlation_filter(df, 'event')
X_pearson = df[pearson_features]
y = df['event']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_pearson, y, test_size=0.2, random_state=42)

# Applying the Wrapper Method of Recursive Feature Elimination (RFE)
rfe_model = RandomForestClassifier(n_estimators=100, random_state=42)
rfe = RFE(estimator=rfe_model, n_features_to_select=5, step=1)
rfe.fit(X_train, y_train)
rfe_features = X_train.columns[rfe.support_].tolist()

X_train_rfe = X_train[rfe_features]
X_test_rfe = X_test[rfe_features]

# Using the First Embedded Method of LASSO
lasso = LogisticRegression(penalty='l1', solver='liblinear', C=1.0, random_state=42)  # approprate filtering using parameter C
lasso.fit(X_train_rfe, y_train)
lasso_model = SelectFromModel(lasso, prefit=True)
lasso_features = X_train_rfe.columns[lasso_model.get_support()].tolist()

X_train_lasso = X_train_rfe[lasso_features]
X_test_lasso = X_test_rfe[lasso_features]

# Using the Second Embedded Method of Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_lasso, y_train)
rf_model = SelectFromModel(rf, prefit=True)
rf_features = X_train_lasso.columns[rf_model.get_support()].tolist()

# Final selected features
X_train_final = X_train_lasso[rf_features]
X_test_final = X_test_lasso[rf_features]

# Print results
print("Original features:", df.columns.tolist())
print("After Pearson Correlation:", pearson_features)
print("After RFE:", rfe_features)
print("After LASSO:", lasso_features)
print("Final features (after Random Forest):", rf_features)

# Evaluating the final model
final_model = RandomForestClassifier(n_estimators=100, random_state=42)
final_model.fit(X_train_final, y_train)


```

```python
# Evaluating the final model
final_model = RandomForestClassifier(n_estimators=100, random_state=42)
final_model.fit(X_train_final, y_train)
accuracy = final_model.score(X_test_final, y_test)
print(f"Final model accuracy: {accuracy:.4f}")
```
